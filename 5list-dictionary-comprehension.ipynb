Q. Filter out string with length greater than 2 and convert them to upper case

```python
string = ['a', 'as' , 'bus', "car", 'dove', 'python']
```

```python
#method 1

result= []
for i in string:
    if len(i)>2:
        result.append(i.upper())
print(result)
```

```python
#method 2

[i.upper() for i in string if len(i)>2]
```

----------------Nested List Comprehension----------------------------


```python
data= [['john','emily', 'micheal', 'mary', 'steven'],
       ['maria', 'juan', 'javiar', 'natalia', 'pillar']]
print(data)
```

Q. produce single list containing all names with two or more'a' in them

```python
#method1
output=[]
for i in data:
    for j in i:
        if j.count('a')>=2:
            output.append(j)
print(output)         
```

```python
#method2
output=[]
for i in data:
    a=[j for j in i if j.count('a')>=2]
    output.extend(a)
print(output)
```

```python
#method3
[j for i in data for j in i if j.count('a')>=2]
```

```python
states= [" Albama ", "Georgia!", "georgia", "Georgia", "FlOrida", "south carolina##", "West virginia?"]

import re #regular expressions

def clean_strings(strings):
    result=[]
    for value in strings:
        value=value.strip()           #remove blank spaces from extreme left and right
        value=re.sub("[!#?]", "", value)    #substitute special character
        value=value.title()              #capitalize first letter of each word
        result.append(value)
    return result

clean_strings(states)
    
```

-------------------- Lambda Functions -------------------------

```python
#small single line anonymous functions returns the result
```

```python
def shortfunc(x):       #method 1
    return x*2

shortfunc(20)
```

```python
equiv = lambda x:x*2      #method 2
equiv(20)
```

```python
equiv = lambda x,y: x*y*2
equiv(20,40)
```

```python
def applylist(l, f):
    return [f(x) for x in l] #item aappended is the output of the function

ints=[4,0,1,5,6]
applylist(ints, lambda x:x*2)  #function is defined through lambda function
```

```python
def attempt_float(x):
    try:
        return float(x)
    except:
        return x    #the code in except block will only be executed if try block raises an exception
```
