# SVC **MODEL**

```python
from sklearn.svm import SVC

from sklearn.model_selection import TimeSeriesSplit, GridSearchCV
```

```python
x_smote,y_smote
```

```python
svc_model= SVC(random_state=32)
svc_model
```

```python
tscv = TimeSeriesSplit(n_splits=3)
```

```python
svc_model.get_params().keys()
```

```python
param_grid = {'kernel' : ['linear', 'rbf',  'sigmoid'],
              'C': [0.001, 0.01, 0.1],
              'gamma': [ 0.1, 0.01]}
svc_gridsearch = GridSearchCV(svc_model, param_grid, verbose=3, cv=tscv, n_jobs=4, scoring='accuracy')
```

```python
svc_gridsearch
```

```python
svc_gridsearch.fit(x_smote,y_smote)
```

```python
svc_results = pd.DataFrame(svc_gridsearch.cv_results_['params'])
```

```python
svc_results['test_score'] = svc_gridsearch.cv_results_['mean_test_score']
```

```python
svc_results
```

```python
import numpy as np

mean_fit_time= svc_gridsearch.cv_results_['mean_fit_time']
mean_score_time= svc_gridsearch.cv_results_['mean_score_time']
n_splits  = svc_gridsearch.n_splits_ #number of splits of training data
n_iter = pd.DataFrame(svc_gridsearch.cv_results_).shape[0] #Iterations per split

print(np.mean(mean_fit_time + mean_score_time) * n_splits * n_iter/60)
```

```python
print(svc_gridsearch.best_params_)
```
